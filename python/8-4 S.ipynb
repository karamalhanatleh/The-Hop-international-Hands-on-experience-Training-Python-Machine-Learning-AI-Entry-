```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
sv=[]
for i in x:
    sv.append(np.square(i))
```

```python
cv=[]
for i in x:
    cv.append(i**3)
```

```python
plt.plot(x , '-.')
plt.show()
```

```python
plt.plot(x,'--')
plt.plot(sv,'-*')
plt.plot(cv,'-^')
plt.legend()
plt.grid(True)
plt.show()
```
