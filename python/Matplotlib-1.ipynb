**Matplotlib**

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Create data for plot

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]

squaredValues = [n ** 2 for n in x] 

cubicValues =  [n ** 3 for n in x]


print("X : ", x)
print("squared values : ", squaredValues)
print("cubic values : ", cubicValues)
```

# Pie Chart

```python
# Create data to represent in Pie Chart

t_shirt_sizes = ['small', 'meduim', 'larg', 'X larg']
num_t_shirts = [45, 36, 75, 70]
```

```python
figure_size = (10,10)
figure_dpi = 90
n_rows = 2
n_columns = 1

fig = plt.figure(figsize= figure_size, dpi = figure_dpi)
ax1 = fig.add_subplot(n_rows, n_columns,1)

ax1.pie(x=num_t_shirts, labels=t_shirt_sizes)


plt.show()
```

```python
#  auto-labeling the percentage


figure_size = (10,10)
figure_dpi = 90
n_rows = 2
n_columns = 1

fig = plt.figure(figsize= figure_size, dpi = figure_dpi)
ax1 = fig.add_subplot(n_rows, n_columns,1)

# ax1.pie(x=num_t_shirts, labels=t_shirt_sizes, autopct='%1.1f%%')

_, _, autotexts = ax1.pie(x=num_t_shirts,labels =t_shirt_sizes, autopct='%1.1f%%',pctdistance=0.8,startangle=90)
for autotext in autotexts:
    autotext.set_color('white')

plt.show()
```

```python
# offsetting a slice with "explode"

figure_size = (10,10)
figure_dpi = 90
n_rows = 2
n_columns = 1

fig = plt.figure(figsize= figure_size, dpi = figure_dpi)
ax1 = fig.add_subplot(n_rows, n_columns,1)


explode = (0, 0, 0.1, 0)  # only "explode"  (i.e. 'larg')

ax1.pie(x=num_t_shirts, labels=t_shirt_sizes, autopct='%1.1f%%',
        explode=explode)


plt.show()
```

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]

farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()

fig.set_dpi(110)
im = ax.imshow(harvest)

# Show all ticks and label them with the respective list entries
plt.xticks(np.arange(len(farmers)), labels=farmers)
plt.yticks(np.arange(len(vegetables)), labels=vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
plt.show()
```
