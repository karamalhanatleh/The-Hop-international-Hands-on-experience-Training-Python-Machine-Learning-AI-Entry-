**correlation coefficient**



*   The correlation coefficient is an equation that is used to determine 

      the strength of the relation between two variables


*   The correlation coefficient always lies between -1 to +1
*    -1 represents X and Y are negatively correlated
*   +1 represents X and Y are positively correlated



---



---


r = n (∑xy) – ∑x ∑y / √ [n* (∑x2 – (∑x)2)] * [n* (∑y2 – (∑y)2)]


*   r = correlation coefficient

*   n = number of observations

*   x = 1st variable in the context



*   y = 2nd variable




---

Absolute Value of r Strength of Relationship


r < 0.3  ==>  None or very weak


0.3 < r <0.5   ==>  Weak


0.5 < r < 0.7  ==>   Moderate


r > 0.7  ==>   Strong

```python
# Solution 6.3
def correlation_coefficient(x, y):
    n = len(x)
    sum_xy = 0
    sum_x = 0
    sum_y = 0
    sum_x2 = 0
    sum_y2 = 0
    
    for i in range(n):
        sum_xy += x[i] * y[i]
        sum_x += x[i]
        sum_y += y[i]
        sum_x2 += x[i] ** 2
        sum_y2 += y[i] ** 2
        
    numerator = (n * sum_xy) - (sum_x * sum_y)
    denominator = ((n * sum_x2) - (sum_x ** 2)) * ((n * sum_y2) - (sum_y ** 2))
    denominator = denominator ** 0.5
    
    return numerator / denominator

total_rooms = df["total_rooms"].values
median_house_value = df["median_house_value"].values

# calculate the correlation coefficient using the function
corr_coef = correlation_coefficient(total_rooms, median_house_value)
corr_coef
```
